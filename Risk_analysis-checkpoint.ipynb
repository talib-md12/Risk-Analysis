## There are many attiributes having more than 40% data missing

```python
null_df = round((application.isnull().sum()/application.shape[0]) * 100 ,2 )
```

```python
plt.figure(figsize = (20,8))
ax = sns.pointplot(data = null_df)
plt.ylabel('Null Value Percentage')
ax.axhline(40, ls='--',color='red',label='Threshold (40%)')
ax.legend()
plt.title('Null Value Percentage')
plt.xticks(rotation = 90 , fontsize = 7 )
plt.grid(axis = 'y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (15,8))
ax = sns.pointplot(data = null_df)
plt.xlabel('columns')
plt.ylabel('null value percentage')
ax.axhline(y= 40 , color = 'red' , label = 'threshold = 40%')
plt.xticks(rotation = 90 ,fontsize = 7)
plt.legend()
plt.tight_layout()

plt.savefig('null_values.png')
plt.show()
```

# there are many attributes having more than 40% missing Values

```python
unwanted_applications = null_df[null_df > 40]
```

```python
noise_cols = unwanted_applications.index
```

```python
null_df2 = round((previous_application.isnull().sum()/ previous_application.shape[0] ) * 100 ,2)
```

```python
plt.figure(figsize=(15,8))
ax = sns.pointplot(data = null_df2)
plt.xlabel('columns')
plt.ylabel('Null value percentage')
plt.xticks(rotation = 45 , fontsize = 8 )
ax.axhline(y = 40 ,label = 'Threshold = 40%' , color = 'red')
plt.legend()
plt.savefig('null_df2.png')
plt.show()
```

```python
unwanted_cols2 = null_df2[null_df2 > 40]
```

```python
application.drop(columns = noise_cols , inplace = True)
```
